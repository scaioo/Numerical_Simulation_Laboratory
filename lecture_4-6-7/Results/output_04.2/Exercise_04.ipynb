#### Statistical mechanics:
The goal of statistical mechanics is to predict the equilibrium macroscopic properties of bodies on the basis of their microscopic structure or laws. The system is described in the phase space of $6N$ dimensions where $N$ is the number of particles and the coordinates $(q,p)=\{(q_i,p_i)\}_{i=1,...3N}$ evolves following the canonical equation of motions.
The problem is that in such type of systems the initial microstate of a macroscopic system is unknowable. Only a macroscpoic variables are measurable and so is preferred a probabilistic description.
Considering a large number of systems with different microscopic features but with the same macrostate. Under ordinary circumstances we expect that the average behaviour of any system (**ensemble**) is identical with the time averaged behaviour of the given system.
Statistical mechanics is based on the **Gibbs ensemble concept**: many microscopic configurations lead to the same macroscopic properties, so it is not necessary to know the deterministic motion of every particle in a system in order to predict its properties. It is sufficient to average over a large number of identical systems, each in a different microscopic configuration, so the macroscopic observables of a system are formulated in terms of ensemble averages. $$ \left<f\right>_{p(q,p)}=\int d^{3N}qd^{3N}pf(q,p)p(q,p)$$However in most experiments physicists perform measurement during a certain time interval. 
- For the **ergodic hypothesis** can be assumed a one-to-one correspondence between the initial phase space coordinates of a system and those that specify the state of the system at a later time: $$ \left<f\right>_{p(q,p)}=\lim_{N_{init} \rightarrow \infty}\left<\left<f\right>_{exp.}\right>_{init.}=\lim_{\Delta t \rightarrow \infty} \left<f\right>_{exp.}$$ where $$\left<f\right>_{exp.}=\frac{1}{\Delta t}\int_{t0}^{t0+\Delta t}dt f(q(t),p(t))$$ is the time average measured in an experiment and $\Delta t \gg 1$, while $$\left<\left<f\right>_{exp.}\right>_{init.}=\frac{1}{N_{init.}}\sum_{j=1}^{N_{init.}}\frac{1}{\Delta t}\int_{t0}^{t0+\Delta t}dt f(q(t,j),p(t,j))=\left<f\right>_{p(p,q)}$$
#### Molecular Dinamics: 
The more difficult and interesting the problem, the more desirable it becomes to have exact results available, both to test existing approximation methods and to point the way towards new approaches. Molecular dynamics (MD) is the science of simulating the time dependent behavior of a system of particles. The time evolution of the set of interacting atoms is followed by integrating their equation of motion with boundary conditions appropriate for the geometry or symmetry of the system. In order to calculate the microscopic behavior of a system from the laws of classical mechanics, MD requires, as an input, a description of the interaction potential.
Any two classical trajectories which are initially very close will eventually diverge from one another exponentially with time (**Lyapunov instability**). Presumably, both the reference trajectory and the perturbed trajectory are diverging from the true solution of Newtonâ€™s equations. Lyapunov instability would seem to be a devastating blow to the whole idea of Molecular Dynamics simulations. The reason why it'not a real problem is that cause micro-canonical ensemble is used to perform simulation , there's no need of exact classical trajecories. The point is that the particle
trajectories must stay on the appropriate constant-energy hypersurface in phase space.
#### Integration algorithm:
Energy conservation is an important criterion, but actually we should distinguish two kinds of energy conservation, namely, short time and long time. Energy conservation is degraded as the time step is increased. The potential energy is a function of the atomic positions (3N) of all the atoms in the system. Due to the complicated nature of this function, there is no analytical solution to the equations of motion and these equation must be solved numerically. The **Verlet algorithm** is the most widely used third order Taylor expansion, starting from the following equations: $$ r(t+\delta t)=r(t)+\delta tv(t)+\frac{1}{2}\delta t^2a(t)+\frac{1}{6}\delta t^3\dot{a}(t)+O(\delta t^4)$$ $$ r(t-\delta t)=r(t)-\delta tv(t)+\frac{1}{2}\delta t^2a(t)-\frac{1}{6}\delta t^3\dot{a}(t)+O(\delta t^4)$$ and the two Taylor expansions we obtain: $$ r(t+\delta t)\approx 2r(t)-r(t-\delta t)+\delta t^2a(t)$$ 
#### Periodic Boundary Conditions:
Because of the big number of particles in the system is impossible to simulate all of them $N\approx 10^23$. Is preferred to use Periodic Boundary Conditions (PDB) to a subset of the system of volume $V$. the cubic box is replicated throughout space to form an infinite lattice and in the course of the simulation, as a molecule moves in the original box, its periodic image in each of the neighbouring boxes moves in exactly the same way. Thus, as a molecule leaves the central box, one of its images will enter through the opposite face.
#### Virial Theorem:
From *generalized equipartition* $$\left< q_k\frac{\partial H}{\partial q_k}\right>=k_BT \Rightarrow -\frac{1}{3} \left< \sum_{i=1}^N \vec{r}_i \vec{\nabla}_iV \right> = -NK_BT$$ Taking into account that: $-\vec{\nabla}_iV=\vec{F}_i^{tot}=\vec{F}_i^{int}+\vec{F}_i^{ext}$ where the virial $W=\frac{1}{3}\sum_{i=1}^Nr_iF_i^{int}$ and considering a pairwise inter-particle interaction $\vec{F}_{ij}=-\vec{F}_{ji}$ can be obtained:
$$ P=\rho K_BT+\frac{1}{3V}\left<\sum\sum_{i<j=1^{N}}|\vec{r}_{ij}|\frac{dV_i}{dr}\right>$$ that in the case of lennard jones interaction is the Pressure written in the exercise above.
### Problem's approach:
The NSL_SIMULATOR perform MD simulations taking in input some parameters from an input file, evolving the start configuration of the particles or spins and measurng some quantities that are returned in an output file.
An example ofcode where a MD is performed is the following:
```cpp
    System SYS;
    SYS.initialize(input_path,output_path); 
    SYS.initialize_properties(input_path,output_path);
    SYS.block_reset(0);
    for(int i=0; i < SYS.get_nbl(); i++){ //loop over blocks
      for(int j=0; j < SYS.get_nsteps(); j++){ //loop over steps in a block
        SYS.step();
        SYS.measure();
        if(j%10 == 0){
          nconf++;
        }
      }
      SYS.averages(output_path,i+1);
      SYS.block_reset(output_path,i+1);
    }
    SYS.finalize(output_path);
  }
```
In the following paragraph the NSL_SIMULATOR's methods will be explained:\

`Initialize`: set the random number generator's seed from the *Primes* file, write the firs line of the output file *acceptance.dat* and start reading the parameters from the *input.dat* file. In particluar read the following properties: 
- simulation type (NVE,NVT;Metropolis or Gibbs),
- restart (to restart from a precedent equilibration)
- set the starting temperature
- set the number of particles or the spins orientations randomly
- set the density, the bolume and the size of the box
- set the cut off radius for the interactions
- set the $\Delta t$ to update particle positions or to accept or reject a move
- read the initial configuration of particles or spins from the configuration file
- Initialize the velocities and the first position updat from a file or following a gaussian distribution

`Initialize properties`: Passing through each property, an output file is initialized for each property measured, a vector that counts the number of properties measured is updated step by step and a bool variable is set to true if such property is measured.

`block reset`: Every time a block is completed this function is called. On the *output.dat* file is written "block completed" and the block average vector is set to zero to measure the next block.

`get nbl`: return the number of block that want to be measured in a simulation.

`get nsteps`: return the number of steps that want to be performed in each block.

`step`: if the simulation is NVE perform a Verlet step, while if the simulation is different perfom a Monte Carlo move sampled with gibbs or metropolis depending on the type of the simulation for the 1D Ising Model or for the Lennard Jones system.

`measure`: measure each property for every step and update the block accumulator vector. 

`averages`: At the end of the block is computed the average over the block for each property measured and is performed the blocking method for each quantity.

`finalize`: write the final configuration at the end of the simulation and save the random seed.

To measure the pressure in the MD simulation in the `measure` function has been implemented an estimation of the virial over the pair interaction of the particles. Calculated the virial is computed the pressure from the formula above and stored over the *_measurement* vector.
In the end in the `averages` funcion has been implemented the calculus of the block average over the pressure.

To avoid this problem i have performed some different simulations and, after an equilibration time, i've saved the final configurations of the particles, the velocities and the final temperature for the solid, liquid and gas phase.

### Code:
To change the parameters on *input.dat* i have written a `change_file` function that change in a file a property:
```cpp
void change_file(const string &filename,string keyword,string new_value){
    ifstream ifile(filename);
    if (!ifile.is_open()) {throw invalid_argument("Impossibile aprire il file per la lettura.");}
    vector<string> lines;
    string line;
    while (getline(ifile, line)) {
        lines.push_back(line);
    }
    ifile.close();
    for (string& line : lines) {
        istringstream iss(line);
        string key;
        string value;
        if (iss >> key >> value) {
            if (key == keyword) {
                line = keyword + " " + new_value;
            }
        }
    }
    ofstream ofile(filename);
    if (!ofile.is_open()) {throw invalid_argument("Impossibile aprire il file per la scrittura.");}
    for (const string& line : lines) {
        ofile << line << endl;
    }
    ofile.close();
}
```
thanks to this function and to `create_directory` of `filesystem` i have created the *equilibration.cpp* code that set the parameters and saves the output in a new directory.
